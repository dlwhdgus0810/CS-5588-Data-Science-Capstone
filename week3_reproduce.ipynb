```python
from easy_control import CtrlX

ctrl_x = CtrlX(
    pretrained_model_name_or_path="stabilityai/stable-diffusion-xl-base-1.0",
    variant="fp16",
    device="cuda",
)
result = ctrl_x.sample(
    structure_image="./test_images/normal_house.png",
    structure_prompt="one of the most expensive houses sold in the US last month",
    appearance_image="./test_images/damaged_house.jpeg",
    appearance_prompt="collapsed houses in Florida due to a strong tornado",
    prompt="a house severely damaged by a tornado, falling trees, dark ocean, broken glasses ",
    positive_prompt="high quality",
    negative_prompt="ugly, blurry, dark, low res, unrealistic",
)
result.show()
```
